## Import dependency

```python
from dotenv import load_dotenv
from IPython.display import display

from uniflow.client import Client
from uniflow.flow.flow_factory import FlowFactory
from uniflow.config import OpenAIConfig
from uniflow.viz import Viz
from uniflow.schema import Context

load_dotenv()
```

```python
FlowFactory.list()
```

## Run model flow to generate data

```python
contents = ["It was a sunny day and the sky color is blue.", "My name is bobby and I am a talent software engineer working on AI/ML."]
data = [
    Context(context=c)
    for c in contents
]
```

```python
config = OpenAIConfig()
client = Client(config)
```

```python
output = client.run(data)
```
